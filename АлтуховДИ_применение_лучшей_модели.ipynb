# Эти строки удалить

```python
list_to_delete = test[(test['native-country'] == '?') | (test['occupation'] == '?') | (test['workclass'] == '?')]

indexes_to_delete = list_to_delete.index

display(list_to_delete)
display(indexes_to_delete)
```

```python
train['native-country'] = train['native-country'].replace('?', np.nan)
train['workclass'] = train['workclass'].replace('?', np.nan)
train['occupation'] = train['occupation'].replace('?', np.nan)

test['native-country'] = test['native-country'].replace('?', np.nan)
test['workclass'] = test['workclass'].replace('?', np.nan)
test['occupation'] = test['occupation'].replace('?', np.nan)

# удаляем строки с вопросами
train.dropna(how='any', inplace=True)
display(train.head())
print(train.shape )

test.dropna(how='any', inplace=True)
display(test.head())
print(test.shape)
```

```python
train['income'] = train['income'].map({"<=50K": 0, ">50K": 1}).astype(int)
```

```python
from category_encoders import TargetEncoder
encoder = TargetEncoder()
features = ['sex', 'race', 'marital-status', 'workclass', 'education', 'occupation', 'relationship', 'native-country']
def target_enc(train, test, features):
    targ_enc = TargetEncoder(cols=features)
    targ_enc.fit(train[features], train['income'])
    
    train = train.join(targ_enc.transform(train[features]).add_suffix('_targ'))
    test = test.join(targ_enc.transform(test[features]).add_suffix('_targ'))
    
    train = train.drop(features, axis=1)
    test = test.drop(features, axis=1)
    
    return train, test
train, test = target_enc(train, test, features)
```
